```python
col_names = ['Kecepatan pelayanan', 'Kemudahan fungsi fungsi untuk digunakan', 
             'Kemudahan nama situs untuk diingat', 'Kemudahan situs untuk dipahami',
             'Kemudahan untuk dihubungi', 'Kemudahan untuk dipahami',
             'Kenyamanan dalam mempelajari materi', 'Keramahan petugas',
             'Manfaat terhadap tugas','Pengetahuan terhadap situs',
            'Sistematika Penulisan','Tampilan situs','Final Score','Klasifikasi']

for col in col_names:
    df_t2[col] = pd.to_numeric(df_t2[col], errors='coerce')
```

```python
df_t2.info()
```

# Proses Membangun Model

```python
feature_names = ['Kecepatan pelayanan', 'Kemudahan fungsi fungsi untuk digunakan', 
             'Kemudahan nama situs untuk diingat', 'Kemudahan situs untuk dipahami',
             'Kemudahan untuk dihubungi', 'Kemudahan untuk dipahami',
             'Kenyamanan dalam mempelajari materi', 'Keramahan petugas',
             'Manfaat terhadap tugas','Pengetahuan terhadap situs',
            'Sistematika Penulisan','Tampilan situs','Final Score']
feature_names2=['Klasifikasi']

x = df_t2[feature_names] # df.drop('label')
y = df_t2[feature_names2]
```

```python
x.head()
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=420)
```

```python
dtClassifier = DecisionTreeClassifier()

dtClassifier.fit(x_train, y_train)

y_pred = dtClassifier.predict(x_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
dtClassifier = DecisionTreeClassifier(criterion='entropy', max_depth=3)

dtClassifier.fit(x_train, y_train)

y_pred = dtClassifier.predict(x_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

Sesuai Service Level Agreement (SLA) Perusahaan, bahwa nilai setiap aspek harus di atas 85

Apabila semua aspek di atas 85, maka dilabeli 1 dan sesuai dengan SLA perusahaan

Apabila ada 1 saja aspek yang di bawah 85, maka dilabeli 0 dan tidak sesuai dengan SLA perusahaan. Sehingga harus dilakuan improvement agar nilai di atas 85 untuk aspek tersebut

pada test 1, dilabeli 1 karena semua aspek mendapatkan nilai di atas 85

pada test 2 dan test 3, dilabeli 0 karena ada aspek yang nilainya di bawah 85
