```python
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_excel('./dataset/Nilai_LIM_1_Pelatihan.xlsx', sheet_name='LIM 1')
```

```python
df
```

```python
dft=df.transpose()
dft
```

```python
x=dft.iloc[18:22]
X=x.transpose()
X
```

```python
X['UBPP Inst']
```

```python
X['UBPP Inst'].isnull()
```

```python
X['UBPP Inst'].fillna(0,inplace=True)
X['UBPP Sarana'].fillna(0,inplace=True)
X['UBPP Penyelenggaraan'].fillna(0,inplace=True)
X['UBPP Materi'].fillna(0,inplace=True)
```

```python
X
```

```python
Y=df['Net Promotor Score']
Y
```

```python
transformer = PolynomialFeatures(include_bias=False, degree=2, interaction_only=False,order='C')
```

```python
transformer.fit(X)
```

```python
X_=transformer.transform(X)
```

```python
X_.shape
```

```python
print(X_)
```

```python
model = LinearRegression()
model.fit(X_,Y)
```

```python
r_sq = model.score(X_,Y)
print(r_sq)
print(model.intercept_)
print(model.coef_)
```

```python
y_pred = model.predict(X_)
```

```python
y_pred
```

```python
model.predict(transformer.transform([[80,80,80,80]]))
```
